```python
import qiskit
import qiskit_aer
from qiskit import QuantumCircuit, ClassicalRegister, QuantumRegister
from qiskit.visualization import plot_histogram
import numpy as np

```

```python
# Step 1: Define quantum and classical registers
q = QuantumRegister(4, 'q')  # Four qubits representing four coins
c = ClassicalRegister(4, 'c')  # Classical bits for measurement
qc = QuantumCircuit(q, c)
```

```python
# Step 2: Apply Hadamard gates to create superposition
qc.h(q)  # Puts all qubits in an equal superposition of real and counterfeit states

# Step 3: Oracle - Fake coin checker (assume q[2] is counterfeit)
qc.cz(q[2], q[3])  # Sample phase flip to mark counterfeit coin

# Step 4: Grover Diffusion (Amplification)
qc.h(q)  # Apply Hadamard again
qc.x(q)  # Flip 0s to 1s and 1s to 0s
qc.h(q[3])  # Apply Hadamard to last qubit (controlled phase shift)
qc.mcx(q[:3], q[3])  # Multi-control X gate (reflection)
qc.h(q[3])
qc.x(q)
qc.h(q)

# Step 5: Measure the qubits
qc.measure(q, c)
qc.draw()
```

```python

# Simulate the circuit
simulator = qiskit_aer.Aer.get_backend('qasm_simulator')
program = qiskit.transpile(qc, simulator)
job = simulator.run(program)

result = job.result()
counts = result.get_counts()

# Show results
plot_histogram(counts)
print(counts)
```
